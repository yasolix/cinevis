```python
def displayASL(asl):
    plt.figure(figsize=(10,10))
    plt.axis([1905,2017,0,30])
    plt.grid(True)
    plt.scatter(asl['Year'],asl['ASL'])

plt.figure(figsize=(20,20))
plt.axis([1915,2017,0,30])
plt.rcParams['xtick.labelsize']=8
plt.rcParams['ytick.labelsize']=8
plt.grid(True)
plt.scatter(newasl['Year'],newasl['ASL'])
plt.plot(meannewasl,'r',linewidth=4.0)
plt.xlabel('Years')
plt.ylabel('Average Shot Length (seconds)')
plt.title('ASL of Movies in Years ')
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.show()
```
